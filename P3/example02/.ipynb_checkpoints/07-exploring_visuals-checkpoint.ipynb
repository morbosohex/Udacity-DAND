# Exploring with Visuals
Use `clean_08.csv` and `clean_18.csv`

```python
# load datasets
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df_08 = pd.read_csv('clean_08.csv')
df_18 = pd.read_csv('clean_18.csv')
```

```python
df_08['mpg'] = (df_08['city_mpg'] + df_08['hwy_mpg'] + df_08['cmb_mpg']) /3
```

```python
df_18['mpg'] = (df_18['city_mpg'] + df_18['hwy_mpg'] + df_18['cmb_mpg']) /3
```

```python
df_08['greenhouse_gas_score'].plot(kind='hist');
```
